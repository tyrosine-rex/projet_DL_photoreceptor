```python
vignettes_dir = '../../res/vignettes_v1_triees'
target_dir = '../../res/vignettes_v1_triees_png'
```

```python
import pathlib
data_dir = pathlib.Path(vignettes_dir)
```

```python
image_count = len(list(data_dir.glob('*/*.tif')))
print(image_count)
```

```python
images = []
labels = []
for path in list(data_dir.glob('*/*.tif')):
    # os.path.dirname(path)[len(os.path.dirname(path))-1]
    # os.path.dirname(path).split('\\')[-1:][0]
    label = os.path.dirname(path)[len(os.path.dirname(path))-1]
    labels.append(label)
    #print(type(path))
    img = tf.keras.utils.load_img(
        path,
        color_mode="grayscale",
        target_size=(60,60),
        interpolation="nearest",
        keep_aspect_ratio=True,
    )
    images.append(img)
```

```python
x = len(images)
n = m.ceil(m.sqrt(x))

for i in range(len(images[:100])):
    ax = plt.subplot(10, 10, i + 1)
    plt.imshow(images[i])
    plt.axis("off")
```

```python
training_ds, validation_ds = tf.keras.utils.image_dataset_from_directory(
    vignettes_dir,
    labels="inferred",
    label_mode="int",
    class_names=None,
    color_mode="grayscale",
    batch_size=64,
    image_size=(60, 60),
    shuffle=True,
    seed=1,
    validation_split=0.2,
    subset="both",
    interpolation="bilinear", # bilinear, nearest, bicubic, area, lanczos3, lanczos5, gaussian, mitchellcubic.
    follow_links=False,
    crop_to_aspect_ratio=True
)
```

## Augmentation des données

```python
data_augmentation = tf.keras.Sequential([
  layers.RandomFlip("horizontal_and_vertical"),
  layers.RandomRotation(0.2),
  layers.RandomContrast(0.5, seed=1)
])
```

```python
# Add the image to a batch.
image = tf.expand_dims(image, 0)
```

```python
plt.figure(figsize=(10, 10))
for i in range(9):
  augmented_image = data_augmentation(image)
  ax = plt.subplot(3, 3, i + 1)
  plt.imshow(augmented_image[0])
  plt.axis("off")
```
